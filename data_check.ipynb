```python
import pandas as pd
```

```python
df = pd.read_csv('user_data.csv')
df.head()
```

```python
df['total_kids'] = df['kids'] + df['teenagers']
```

```python
df['age'].value_counts().sort_index().plot(kind='bar', figsize=(15,5), xlabel='Возраст', ylabel='Количество пользователей', title='Распределение возраста пользователей')
print('Средний возраст', df['age'].mean())
print('Медианный возраст', df['age'].median())
```

```python
df.groupby('age').income.mean().plot(kind='bar', figsize=(15, 5), xlabel='Возраст', ylabel='Средний доход', title='Средний доход по возрасту')
```

```python
df.education.value_counts().plot(kind='pie', title='Распределение образования')
```

```python
df.marital.value_counts().plot(kind='pie', title='Распределение семейного статуса')
```
